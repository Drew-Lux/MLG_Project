## Import all Modules

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

## Load data and make a copy

```python
data = pd.read_csv('../data/Diabetes_and_LifeStyle_Dataset_.csv')
df = data.copy()
```

**Summary of data**

```python
print("--- Dataset Info ---")
print(df.info())
print("\n--- Summary Statistics ---")
display(df.describe())
```

**Show first 5 rows**

```python
print("\n--- First 5 Rows ---")
display(df.head())
```

**Removing all the null values**

```python
print("--- Null Values per Column ---")
print(df.isnull().sum())


df_clean = df.dropna().copy()


print(f"\nOriginal rows: {len(df)}")
print(f"Rows after cleaning: {len(df_clean)}")
print(f"Total rows removed: {len(df) - len(df_clean)}")
```

## Graph Diabetes Distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='diabetes_stage', palette='viridis')
plt.title('Distribution of Diabetes Stages')
plt.show()
```

## Mapping and Correlation

```python
mappings = {
    'education_level': {'Highschool': 0, 'Graduate': 1, 'Post-Graduate': 2},
    'income_level': {'Low': 0, 'Lower-Middle': 1, 'Middle': 2, 'High': 3},
    'employment_status': {'Unemployed': 0, 'Employed': 1, 'Retired': 2},
    'smoking_status': {'Never': 0, 'Former': 1, 'Current': 2},
    'diabetes_stage': {'No Diabetes': 0, 'Pre-diabetes': 1, 'Type 2': 2}
}

# Apply the manual mappings
for col, mapping in mappings.items():
    if col in df.columns:
        df[col] = df[col].map(mapping)

# Use Label Encoding for Nominal Data 
le = LabelEncoder()

for col in ['gender', 'ethnicity']:
    if col in df.columns:
        df[col] = le.fit_transform(df[col])


df = df.dropna() 

print("Preprocessing complete. All columns are now numerical.")
print(df.head())
```

**Correlation Matrix**

```python
# Create the correlation matrix
corr = df.corr(numeric_only=True)

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the figure with a larger scale
plt.figure(figsize=(16, 10))

# Draw the heatmap with the mask
sns.heatmap(corr, 
            mask=mask, 
            annot=True, 
            cmap='coolwarm', 
            fmt='.2f', 
            center=0,
            linewidths=.5, 
            cbar_kws={"shrink": .8})

plt.title('Lifestyle & Health Correlation (Cleaned View)', fontsize=16)
plt.show()
```

## Cleaning Data
**Primary Features**

```python
primary_features = [
    'hba1c', 
    'bmi', 
    'systolic_bp', 
    'physical_activity_minutes_per_week', 
    'diet_score', 
    'Age'
]

output_fields = [
    'diabetes_risk_score', 
    'diabetes_stage', 
    'diagnosed_diabetes'
]

# Extract only these columns
df_final = df[primary_features + output_fields].copy()

# Final Clean: Remove any rows with missing data in these specific columns
df_final = df_final.dropna()


print(f"Final dataset shape: {df_final.shape}")
print(df_final.head())
```

**Scaling**

```python
# Initialize the scaler
scaler = StandardScaler()


df_scaled = df_final.copy()
df_scaled[primary_features] = scaler.fit_transform(df_final[primary_features])
```

**Graph Scaled Data**

```python
plt.figure(figsize=(14, 8))
sns.set_theme(style="whitegrid")

sns.boxplot(data=df_scaled[primary_features], palette="Set2")

plt.axhline(0, color='black', linestyle='--', alpha=0.5)

plt.title('Scaled Feature Distributions (Mean=0, Std=1)', fontsize=16)
plt.xticks(rotation=45)
plt.ylabel('Standard Deviations from Mean')
plt.show()
```

**Save cleaned and scaled data**

```python
df_scaled.to_csv('../data/Diabetes_scaled_for_modeling.csv', index=False)
print("Data scaled and saved. Ready for K-Means and XGBoost.")
```
